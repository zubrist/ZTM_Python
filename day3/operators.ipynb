# Arithmetic Operators


```python
# addition
x = 10
y = 20

z= x+y

print(z)
```

```python
# Subtraction
z = x-y
print(z)
```

```python
#multiplication
z = x*y
print(z)
```

```python
# division
z = x/y
print(z)

# floor division
zf = x//y
print(zf)
```

### divide by zero error in python


```python
try:
    z = x/0
except ZeroDivisionError:
    print("You can't divide by zero")
```

```python
# exponentiation
z = x**y
print(z)
```

```python
# modulo
z = x%y
print(z)
```
