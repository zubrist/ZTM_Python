```python
# >>=
y =5
y>>= 1
# y = y >> 1 : means shift the bits of y to the right by 1
'''
5 in binary is 101
shifting to the right by 1 means we drop the last bit and add a 0 at the beginning
so 101 becomes 010 which is 2 in decimal
'''
print(y)
```

```python
# <<=
y <<= 1
# y = y << 1 : means shift the bits of y to the left by 1
'''
2 in binary is 010
shifting to the left by 1 means we drop the first bit and add a 0 at the end
so 010 becomes 100 which is 4 in decimal
'''
print(y)
```

```python
# := (walrus operator)
# allows us to assign a value to a variable as part of an expression

'''
Detailed explanation of the walrus operator:
The walrus operator (:=) is a new assignment expression introduced in Python 3.8. It allows you
to assign a value to a variable as part of an expression, which can be useful in situations 
where you want to use the value immediately after assigning it.

It assigns a value AND returns that value to be used in the expression
   
'''
if (n := len("Hello World")) > 5:
    print("String is too long")
    
```

# Identity Operator

```python
# is | returns True if both operands refer to the same object, otherwise returns False
a = [1, 2, 3]
b = a

a is b
```

```python
# difference between == and is
'''
The == operator checks for value equality, meaning it checks if the values of the operands are
equal. The is operator checks for identity, meaning it checks if both operands refer to the 
same object
'''
# example
a = [1, 2, 3]
b = [1, 2, 3]

a == b  # True because the values are equal
a is b  # False because they are different objects in memory
```

```python
# is not | returns True if the operands do not refer to the same object, otherwise returns False
a is not b  # True because they are different objects in memory
```

# Membership Operators

```python
# in | returns true if a sequence with the specified value is present in the object, otherwise returns False
s = "Hello World"
if "Hello" in s:
    print("Found 'Hello' in the string")


```

```python
# not in | returns True if a sequence with the specified value is not present in the object, otherwise returns False
if "Hi" not in s:
    print("Did not find 'Hi' in the string")
```

# Bitwise Operator

```python
# & AND | Sets each bit to 1 if both bits are 1
a = 5  # in binary: 0101
b = 3  # in binary: 0011
a & b  # in binary: 0001 which is 1 in decimal
```

```python
# | OR / Sets each bit to 1 if one of two bits is 1
a | b  # in binary: 0111 which is 7 in decimal
```

```python
# ^ XOR | Sets each bit to 1 if only one of two bits is 1
a ^ b  # in binary: 0110 which is 6 in decimal
```

```python
# ~ NOT | Inverts all the bits
~a  # in binary: 1010 which is -6 in decimal (two's complement representation)
```

```python
# << left shift | Shifts the bits of the first operand to the left by the number of positions specified by the second operand
a << 1  # in binary: 1010 which is 10 in decimal

b << 2 # in binary: 1100 which is 12 in decimal
```
