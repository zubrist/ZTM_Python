```python
# Subtraction
z = x-y
print(z)
```

# Assignment Operators

```python
# assignment operators
x = 10
```

```python
x += 5
# x = x + 5
print(x)    
```

```python
x -= 5
# x = x - 5
print(x)
```

```python
# %=
x %= 3
# x = x % 3
print(x)
```

```python
# ^=
x ^= 2
# x = x ^ 2
print(x)
```

```python
# >>=
y =5
y>>= 1
# y = y >> 1 : means shift the bits of y to the right by 1
'''
5 in binary is 101
shifting to the right by 1 means we drop the last bit and add a 0 at the beginning
so 101 becomes 010 which is 2 in decimal
'''
print(y)
```

```python
# <<=
y <<= 1
# y = y << 1 : means shift the bits of y to the left by 1
'''
2 in binary is 010
shifting to the left by 1 means we drop the first bit and add a 0 at the end
so 010 becomes 100 which is 4 in decimal
'''
print(y)
```

```python
# := (walrus operator)
# allows us to assign a value to a variable as part of an expression

'''
Detailed explanation of the walrus operator:
The walrus operator (:=) is a new assignment expression introduced in Python 3.8. It allows you
to assign a value to a variable as part of an expression, which can be useful in situations 
where you want to use the value immediately after assigning it.

It assigns a value AND returns that value to be used in the expression
   
'''
if (n := len("Hello World")) > 5:
    print("String is too long")
    
```

# Comparison Operator

```python
# ==  | returns True if the values of the operands are equal, otherwise returns False
x == y
```

```python
# != | returns True if the values of the operands are not equal, otherwise returns False
x != y
```

```python
# grater than >  | returns True if the value of the left operand is greater than the value of the right operand, otherwise returns False
x > y   
```

```python
# less than <  | returns True if the value of the left operand is less than the value of the right operand, otherwise returns False

x < y
```

```python
# greater than or equal to >=  | returns True if the value of the left operand is greater than or equal to the value of the right operand, otherwise returns False
x >= y
```

```python
# less than equal to <=  | returns True if the value of the left operand is less than or equal to the value of the right operand, otherwise returns False
x <= y
```

# Logical Operator

```python
# and | returns True if both operands are true, otherwise returns False
x > 5 and y < 30
```

```python
# or | returns True if at least one of the operands is true, otherwise returns False
x > 5 or y < 30
```

```python
# not | returns True if the operand is false, otherwise returns False
not x > 5
```

```python
not(x > 5 and y < 30)
```
