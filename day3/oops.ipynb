```python
# Inheritance : How it works
"""
Single Inheritance: A child class inherits from a single parent class.
Diagram:

_______________
|Parent Class |
---------------
     |
     |
     |
______________
|Child Class |
--------------

things to note:
- The child class can override methods of the parent class.
- The child class can add new methods and attributes.
- child can access the parent members following ways-  
     1. Using the parent class name: ParentClassName.method(self)
     2. Using the super() function: super().method()

"""
```

```python
#example of single inheritance
class ParentClass:
    def __init__(self, name):
        self.name = name

    def greet(self):
        return f"Hello, {self.name}!"
    
class ChildClass(ParentClass):
    def __init__(self, name, age):
        super().__init__(name)  # Call the parent class constructor
        #or ParentClass.__init__(self, name)
        self.age = age
    def display_info(self):
        return f"{self.name} is {self.age} years old."
child_object = ChildClass("Bob", 30)
print(child_object.greet())  # Inherited method from ParentClass
print(child_object.display_info())  # Method from ChildClass
```

```python
# Multi-level Inheritance
"""
In multi-level inheritance, a child class inherits from a parent class, and then another
child class inherits from that child class.

Diagram:    
_______________
|Parent Class |
---------------
        |
        |
        |
_______________
|Child Class 1|
---------------
        |
        |
        |
_______________
|Child Class 2|
---------------

Things to note:
- Child Class 1 inherits from Parent Class, and Child Class 2 inherits from Child Class 1.
- Child Class 2 has access to the members of both Parent Class and Child Class 1.
- Child Class 1 can override methods of Parent Class, and Child Class 2 can override methods
    of both Parent Class and Child Class 1.
- The child classes can add new methods and attributes as needed.
    
"""

```

```python
# Exaxmple of multi-level inheritance
class ParentClass:
    def __init__(self, name):
        self.name = name

    def greet(self):
        return f"Hello, {self.name}!"
    
    
class ChildClass1(ParentClass):
    def __init__(self, name, age):
        super().__init__(name)  # Call the parent class constructor
        self.age = age
    def display_info(self):
        return f"{self.name} is {self.age} years old."
    
    
class ChildClass2(ChildClass1):
    def __init__(self, name, age, city):
        super().__init__(name, age)  # Call the parent class constructor i.e ChildClass1 constructor
        self.city = city
    def display_full_info(self):
        return f"{self.name} is {self.age} years old and lives in {self.city}."
    
    
child_object = ChildClass2("Charlie", 25, "New York")
print(child_object.greet())  # Inherited method from ParentClass
print(child_object.display_info())  # Inherited method from ChildClass1
print(child_object.display_full_info())  # Method from ChildClass2
```

```python
# Multiple Inherirance 
"""
IN multiple inheritance, a child class can inherit from more than one parent class.
Diagram:
_______________ 
|Parent Class 1|



Things to note:
- The child class can inherit attributes and methods from multiple parent classes.
- If there are conflicting attributes or methods in the parent classes, 
  the child class will use the method resolution order (MRO) to determine which one to use.
- The child class can override methods from any of the parent classes and can also add new 
  methods and attributes as needed.

How MRO works:(in simple terms)
- The MRO is determined using the C3 linearization algorithm, which ensures a consistent 
    method resolution order.
- The MRO is calculated based on the order of the parent classes in the class definition 
    and the hierarchy of the classes.
- The MRO can be accessed using the __mro__ attribute of the class or the mro() method.    

"""
```

```python
#Example of multiple inheritance
class ParentClass1:
    def method1(self):
        return "This is method 1 from Parent Class 1."
    
class ParentClass2:
    def method2(self):
        return "This is method 2 from Parent Class 2."
    
class ChildClass(ParentClass1, ParentClass2):
    def method3(self):
        return "This is method 3 from Child Class."
child_object = ChildClass()

print(child_object.method1())  # Inherited method from ParentClass1
print(child_object.method2())  # Inherited method from ParentClass2
print(child_object.method3())  # Method from ChildClass
```

```python
# the Diamond Problem in Multiple Inheritance
"""
The Diamond Problem occurs in multiple inheritance when a child class inherits from two parent 
classes that both inherit from a common ancestor. 
This can lead to ambiguity in method resolution, as the child class may not know 
which parent class's method to use.
"""

# Example of the Diamond Problem
class A:
    def method(self):
        return "Method from class A"
class B(A):
    def method(self):
        return "Method from class B"
class C(A):
    def method(self):
        return "Method from class C"
class D(B, C):
    pass

d_object = D()
print(d_object.method())  # This will call the method from class B due to MRO

#check the MRO of class D
print(D.__mro__)  # Output: (<class '__main__.D'>, <class '__main__.B'>, <class '__main__.C'>, <class '__main__.A'>, <class 'object'>)
print(D.mro())  # Output: [<class '__main__.D'>, <class '__main__.B'>, <class '__main__.C'>, <class '__main__.A'>, <class 'object'>]
```
