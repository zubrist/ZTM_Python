## what is CORSmiddleware in fastapi 

> the CORSmiddleware is a machenism that sits between the froendend and backend . It let us controll which external origins are allowed to male HTTP request to our fastapi backed .

> It implements Cross-origin Resource Sharing browser security mechanism that allows our apis to talk with fronend from differnt domains.


````python
from fastapi import FastAPI
from fastapi.cors.middleware import CORSMiddleware 


app= FastAPI()

app.add_middleware(
    allow_origins = ["*"],
    allow_credentials=  True,
    allow_methods = ["*"],
    allow_headers = ["*"],
)
````
