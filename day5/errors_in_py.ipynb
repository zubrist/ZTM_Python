## Types of errros in Python
There are several types of errors in Python, including:
1. **Syntax Errors**: These occur when the code is not written in a valid Python syntax. For example, missing a colon at the end of a function definition or using incorrect indentation.
2. **Runtime Errors**: These occur during the execution of the program. For example,
    - **ZeroDivisionError**: When you try to divide a number by zero.
    - **NameError**: When you try to use a variable that has not been defined.
    - **TypeError**: When you try to perform an operation on incompatible data types.
3. **Logical Errors**: These occur when the code runs without throwing any errors, but produces incorrect results. For example, using the wrong operator or incorrect logic in a conditional statement.
4. **Import Errors**: These occur when you try to import a module that does not exist or is not installed in your environment.
5. **File Not Found Errors**: These occur when you try to access a file that does not exist in the specified location.
6. **Value Errors**: These occur when a function receives an argument of the correct type but an inappropriate value. For example, trying to convert a string that does not represent a number into an integer.
7. **Index Errors**: These occur when you try to access an index that is out of range in a list or other sequence.
8. **Key Errors**: These occur when you try to access a key that does not exist in a dictionary.
9. **Attribute Errors**: These occur when you try to access an attribute that does not exist for an object.
10. **Indentation Errors**: These occur when the code is not properly indented, which is crucial in Python for defining code blocks.

 ## Types of error handling
There are several types of error handling techniques that can be used in programming, including:
1. **Try-Catch Blocks**: This is a common method where you wrap code that may throw an error in a try block, and handle the error in a catch block.


```python   
try:
    # Code that may throw an error
    result = 10 / 0
except ZeroDivisionError as e:
    print("Cannot divide by zero:", e)
```
2. **Finally Block**: This block is used to execute code that must run regardless of whether an error occurred or not. It is often used for cleanup actions.

```python
try:
    # Code that may throw an error
    result = 10 / 0
except ZeroDivisionError as e:
    print("Cannot divide by zero:", e)
finally:
    print("This block always executes.")
```
3. **Raising Exceptions**: You can raise exceptions intentionally when a certain condition is met in your code.

```python
def divide(a, b):
    if b == 0:
        raise ValueError("Cannot divide by zero.")
    return a / b    
try:    
    result = divide(10, 0)
except ValueError as e:
    print(e)
```
4. **Custom Exceptions**: You can create your own exception classes to handle specific error conditions in a more descriptive way.

```python
class CustomError(Exception):
    pass

def divide(a, b):
    if b == 0:
        raise CustomError("Cannot divide by zero.")
    return a / b

try:
    result = divide(10, 0)
except CustomError as e:
    print(e)
```
5. **Using Assertions**: Assertions are used to check for conditions that should always be true in your code. If the condition is false, an AssertionError is raised.

```python
def divide(a, b):
    assert b != 0, "Cannot divide by zero."
    return a / b

try:
    result = divide(10, 0)
except AssertionError as e:
    print(e)
```
6. **Context Managers**: These are used to manage resources, such as file handling, ensuring that resources are properly released after use.
```python
with open('file.txt', 'r') as file:
    data = file.read()
```
In this example, the file will be automatically closed after the block of code is executed, even if an error occurs.

7. **Logging**: Instead of just printing error messages, you can use logging to record errors in a more structured way, which can be helpful for debugging and monitoring applications.

```python
import logging
logging.basicConfig(filename='app.log', level=logging.ERROR)
try:
    result = 10 / 0
except ZeroDivisionError as e:
    logging.error("Cannot divide by zero: %s", e)
``` 



## understanding Loggin for all the levels and when and how to use it
Logging is a powerful tool for tracking events that happen when some software runs. It can be used to record information about the execution of a program, which can be helpful for debugging and monitoring applications. Python's built-in logging module provides several levels of logging, each with its own purpose:
1. **DEBUG**: This level is used for detailed information, typically of interest only when diagnosing problems. It is the lowest level of logging and is usually turned off in production environments.

```python   
import logging
logging.basicConfig(level=logging.DEBUG)

#inside a function or method
logging.debug("This is a debug message.")
```

3. **WARNING**: This level is used to indicate that something unexpected happened, or indicative of some problem in the near future (e.g., ‘disk space low’). The software is still working as expected.

```python
import logging

logging.basicConfig(level=logging.WARNING, format="%(levelname)s: %(message)s")

def process_payment(amount: float, balance: float):
    if balance < 1000:
        logging.warning("Account balance is low: %.2f", balance)  # still works, but risky soon
    
    if amount > balance:
        raise ValueError("Insufficient balance")

    return "Payment successful"

print(process_payment(200, 900))
``` 

4. **ERROR**: This level is used to indicate a more serious problem that has occurred. The software has not been able to perform some function.

```python
import logging
logging.basicConfig(level=logging.ERROR, format = "%(levelname)s : %(asctime)s: #%(filename)s: %(lineno)d %(message)s")


def payment_processing(amount: float, balance: float):
    if amount > balance:
        logging.error("Payment failed due to insufficient balance. Attempted amount: %.2f, Available balance: %.2f", amount, balance)
        raise ValueError("Insufficient balance")
    return "Payment successful"
```    


>  Difference between logging an error and raising an exception

***Key difference:***

```logging.error()``` — records the error but continues execution.

```raise ValueError()``` — stops execution and signals an error to the caller.


5. **CRITICAL**: This level is used to indicate a very serious error that has occurred. The software may be unable to continue running.
    
```python
import logging
logging.basicConfig(level=logging.CRITICAL, format = "%(levelname)s : %(asctime)s: #%(filename)s: %(lineno)d %(message)s")

def critical_error_example():
    logging.critical("A critical error occurred. The system may be unable to continue.")
    raise SystemExit("Critical error encountered. Exiting program.")
```
