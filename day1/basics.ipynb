```python
# Floor division rounds towards negative infinity
#5 // 3       # => 1
-5 // 3      # => -2
# why -2 : -5 / 3 = -1.666... and the next lowest integer is -2

5.0 // 3.0   # => 1.0 

-5.0 // 3.0  # => -2.0
```

```python
# The result of division is always a float

10.0 / 3 
```

```python
# difference btwn // and /

# Division (/) - always returns float
10 / 3      # => 3.3333...
10.0 / 3    # => 3.3333...

# Floor division (//) - returns integer (rounds DOWN towards -∞)
10 // 3     # => 3
-5 // 3     # => -2  (not -1, because -1.666... rounds down to -2)
5.0 // 3.0  # => 1.0 (still a float, but rounded down)
```
